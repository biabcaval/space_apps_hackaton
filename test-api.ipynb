````python
import requests
from netrc import netrc
from urllib.parse import urlparse
import os

def download_nasa_file(url, output_file):
    # Configurar autenticação usando .netrc
    urs = 'urs.earthdata.nasa.gov'
    netrc_path = os.path.expanduser('~/.netrc')
    
    try:
        username, _, password = netrc(netrc_path).authenticators(urs)
        
        # Fazer o download
        response = requests.get(url, auth=(username, password))
        if response.status_code == 200:
            with open(output_file, 'wb') as f:
                f.write(response.content)
            print(f"Arquivo baixado com sucesso: {output_file}")
        else:
            print(f"Erro ao baixar: {response.status_code}")
            
    except Exception as e:
        print(f"Erro: {e}")
        print("Certifique-se de ter configurado suas credenciais no arquivo .netrc")

# URL do arquivo
url = "https://data.gesdisc.earthdata.nasa.gov/data/Aqua_AIRS_NRT/AIRS2RET_NRT.7.0/2025/277/AIRS.2025.10.04.178.L2.RetStd_IR.v7.0.9.3.R25277154234.hdf"
output_file = "AIRS.2025.10.04.178.L2.RetStd_IR.v7.0.9.3.R25277154234.hdf"

download_nasa_file(url, output_file)
````

````python
from datetime import datetime, timedelta

def generate_airs_urls(start_date, end_date, granules=range(1, 241)):
    """
    Gera URLs para dados AIRS entre duas datas, considerando os granules de 6 minutos
    
    Args:
        start_date (str): Data inicial no formato 'YYYY-MM-DD'
        end_date (str): Data final no formato 'YYYY-MM-DD'
        granules (range): Range de granules a serem baixados (1-240)
    
    Returns:
        list: Lista de tuplas (url, nome_arquivo)
    """
    base_url = "https://data.gesdisc.earthdata.nasa.gov/data/Aqua_AIRS_NRT/AIRS2RET_NRT.7.0"
    urls = []
    
    current = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')
    
    while current <= end:
        day_of_year = current.timetuple().tm_yday
        
        # Gerar URLs para cada granule do dia
        for granule in granules:
            # Formato: AIRS.YYYY.MM.DD.GGG.L2.RetStd_IR.v7.0.9.3.RXXXXXXXX.hdf
            filename = f"AIRS.{current.year}.{current.month:02d}.{current.day:02d}.{granule:03d}.L2.RetStd_IR.*.hdf"
            url = f"{base_url}/{current.year}/{day_of_year:03d}/{filename}"
            urls.append((url, filename))
        
        current += timedelta(days=1)
    
    return urls

def download_airs_data(start_date, end_date, output_dir="dados_airs", start_granule=1, end_granule=240):
    """
    Baixa dados AIRS para um intervalo de datas
    
    Args:
        start_date (str): Data inicial no formato 'YYYY-MM-DD'
        end_date (str): Data final no formato 'YYYY-MM-DD'
        output_dir (str): Diretório onde os arquivos serão salvos
        start_granule (int): Granule inicial (1-240)
        end_granule (int): Granule final (1-240)
    """
    os.makedirs(output_dir, exist_ok=True)
    
    # Gerar URLs com granules específicos
    urls = generate_airs_urls(start_date, end_date, range(start_granule, end_granule + 1))
    
    total_files = len(urls)
    print(f"Iniciando download de {total_files} arquivos...")
    
    for i, (url_pattern, filename_pattern) in enumerate(urls, 1):
        try:
            print(f"Baixando arquivo {i}/{total_files}: {filename_pattern}")
            download_nasa_file(url_pattern, os.path.join(output_dir, filename_pattern))
        except Exception as e:
            print(f"Erro ao baixar {filename_pattern}: {e}")

# Exemplo de uso
start_date = "2025-10-04"
end_date = "2025-10-04"

# Baixar dados das primeiras 4 horas do dia (40 granules = 4 horas)
download_airs_data(
    start_date=start_date,
    end_date=end_date,
    start_granule=1,
    end_granule=40
)
````

````python
import os
import requests
from datetime import datetime, timedelta
from bs4 import BeautifulSoup

BASE_URL = "https://discnrt1.gesdisc.eosdis.nasa.gov/data/Aqua_AIRS_NRT/AIRS2CCF_NRT.7.0/Recent/"
DOWNLOAD_DIR = "./nasa_data"

os.makedirs(DOWNLOAD_DIR, exist_ok=True)

# Automatically use your ~/.netrc credentials
session = requests.Session()
session.auth = None  # Let requests read from .netrc automatically
session.cookies = requests.cookies.RequestsCookieJar()

def list_files_for_date(date_str):
    """List available files for a given date directory."""
    url = f"{BASE_URL}{date_str}/"
    resp = session.get(url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    files = [a["href"] for a in soup.find_all("a") if a["href"].endswith((".hdf", ".nc4"))]
    return [url + f for f in files]

def download_file(url):
    """Download a single file if not already present."""
    local_name = os.path.join(DOWNLOAD_DIR, os.path.basename(url))
    if os.path.exists(local_name):
        return
    with session.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_name, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)

def main():
    today = datetime.utcnow().date()
    for i in range(7):
        day = today - timedelta(days=i)
        date_str = day.strftime("%Y.%m.%d")
        try:
            print(f"Listing files for {date_str}...")
            files = list_files_for_date(date_str)
            print(f"Found {len(files)} files.")
            for file_url in files:
                print(f"Downloading: {os.path.basename(file_url)}")
                download_file(file_url)
        except requests.HTTPError as e:
            if e.response.status_code == 404:
                print(f"No data for {date_str}.")
            else:
                raise

main()
````

````python
import tempfile
import os
from pyhdf.SD import SD, SDC

def process_airs_data(response):
    """
    Processa os dados AIRS do response e retorna um DataFrame
    
    Args:
        response: Response object do requests
    Returns:
        pd.DataFrame: DataFrame com os dados processados
    """
    # Criar arquivo temporário
    with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
        # Escrever conteúdo do response no arquivo temporário
        tmp_file.write(response.content)
        tmp_file.flush()
        
        try:
            # Abrir arquivo HDF usando o caminho do arquivo temporário
            hdf_file = SD(tmp_file.name, SDC.READ)
            
            # Leitura dos principais poluentes
            o3_total = hdf_file.select('totO3Std').get()
            co_total = hdf_file.select('CO_total_column').get()
            ch4_total = hdf_file.select('CH4_total_column').get()
            
            # Leitura das coordenadas
            latitude = hdf_file.select('Latitude').get()
            longitude = hdf_file.select('Longitude').get()
            
            # Criar DataFrame com os dados
            df = pd.DataFrame({
                'Latitude': latitude.flatten(),
                'Longitude': longitude.flatten(),
                'O3_Total': o3_total.flatten(),
                'CO_Total': co_total.flatten(),
                'CH4_Total': ch4_total.flatten()
            })
            
            return df
            
        except Exception as e:
            print(f"Erro ao processar arquivo: {e}")
            return None
        finally:
            if 'hdf_file' in locals():
                hdf_file.end()
            # Remover arquivo temporário
            os.unlink(tmp_file.name)
````

````python
import tempfile
import os
from pyhdf.SD import SD, SDC

def process_airs_data(response):
    """
    Processa os dados AIRS do response e retorna um DataFrame
    
    Args:
        response: Response object do requests
    Returns:
        pd.DataFrame: DataFrame com os dados processados
    """
    # Criar arquivo temporário
    with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
        # Escrever conteúdo do response no arquivo temporário
        tmp_file.write(response.content)
        tmp_file.flush()
        
        try:
            # Abrir arquivo HDF usando o caminho do arquivo temporário
            hdf_file = SD(tmp_file.name, SDC.READ)
            
            # Leitura dos principais poluentes
            o3_total = hdf_file.select('totO3Std').get()
            co_total = hdf_file.select('CO_total_column').get()
            ch4_total = hdf_file.select('CH4_total_column').get()
            
            # Leitura das coordenadas
            latitude = hdf_file.select('Latitude').get()
            longitude = hdf_file.select('Longitude').get()
            
            # Criar DataFrame com os dados
            df = pd.DataFrame({
                'Latitude': latitude.flatten(),
                'Longitude': longitude.flatten(),
                'O3_Total': o3_total.flatten(),
                'CO_Total': co_total.flatten(),
                'CH4_Total': ch4_total.flatten()
            })
            
            return df
            
        except Exception as e:
            print(f"Erro ao processar arquivo: {e}")
            return None
        finally:
            if 'hdf_file' in locals():
                hdf_file.end()
            # Remover arquivo temporário
            os.unlink(tmp_file.name)

def download_and_process_airs(url, auth):
    """
    Baixa e processa dados AIRS diretamente em memória
    
    Args:
        url (str): URL do arquivo AIRS
        auth (tuple): Tupla com (username, password)
    Returns:
        pd.DataFrame: DataFrame com os dados processados
    """
    try:
        response = requests.get(url, auth=auth, stream=True)
        response.raise_for_status()
        return process_airs_data(response)
    except Exception as e:
        print(f"Erro ao baixar arquivo: {e}")
        return None

# Exemplo de uso
def main():
    urs = 'urs.earthdata.nasa.gov'
    netrc_path = os.path.expanduser('~/.netrc')
    
    try:
        username, _, password = netrc(netrc_path).authenticators(urs)
        auth = (username, password)
        
        url = "https://data.gesdisc.earthdata.nasa.gov/data/Aqua_AIRS_NRT/AIRS2RET_NRT.7.0/2025/277/AIRS.2025.10.04.178.L2.RetStd_IR.v7.0.9.3.R25277154234.hdf"
        
        df = download_and_process_airs(url, auth)
        if df is not None:
            print("\nEstatísticas dos poluentes:")
            print(df.describe())

            # Calcular correlações
            corr = df[['O3_Total', 'CO_Total', 'CH4_Total']].corr()
            print("\nCorrelação entre poluentes:")
            print(corr)

            df.to_csv(f"processed_airs_data.csv", index=False)

    except Exception as e:
        print(f"Erro: {e}")
        print("Certifique-se de ter configurado suas credenciais no arquivo .netrc")

if __name__ == "__main__":
    main()
````
